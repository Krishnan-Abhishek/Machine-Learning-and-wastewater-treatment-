Step4 : Exploratory Data Analysis(EDA)

```python
df.describe() #Description of the data points in that column : count , mean , standard deviation, minimum value, maximum value, 25% percentile, 50% percentile and 75% percentile
```

```python
#Histogram to understand distribution
import warnings
warnings.filterwarnings("ignore")
for i in df.select_dtypes(include="number").columns:
    sns.histplot(data=df,x=i)
    plt.show()
```

```python
#Boxplot-to-identify outliers
import warnings
warnings.filterwarnings("ignore")
for i in df.select_dtypes(include="number").columns:
    sns.boxplot(data=df,x=i)
    plt.show()
```

```python
#scatter plots to understand the relationship
for i in ['CFX [initial concentration of CFX (mg/L)or PPM]',
       'UV (UV/Visible LIGHT)', 'PD (Photocatalytic Doses g/L)',
       'T (Reaction temp)', 'PH (pH Values)', 'C1 (Concentration of NaHCO3)',
       'C2 (Concentration of NaCl)', 'C3 (Concentration of Na2So4)',
       'RT (Reaction time)', 'C (C/C0)']:
    sns.scatterplot(data=df,x = i,y = 'C (C/C0)')
    plt.show()
df.select_dtypes(include="number").columns
```

```python
#s=df.select_dtypes(include = "number").corr()
#sns.heatmap(s,annot=True)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame, and it contains numerical columns
# Calculate the correlation matrix of the numeric columns
s = df.select_dtypes(include="number").corr()

# Create the heatmap
plt.figure(figsize=(10, 8))  # Adjust size if necessary
sns.heatmap(s, annot=True, cmap="coolwarm", linewidths=0.5)

# Save the heatmap to a file (e.g., PNG format)
plt.savefig("correlation_heatmap.png")

# Optionally, display the heatmap
plt.show()
```
